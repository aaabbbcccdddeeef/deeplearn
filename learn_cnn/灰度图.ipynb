```python
import numpy as np
#三维数组的表现形式，表示有2个2维，有2个一维，可以这么理解最外n层里有多少行里面的数组
image = np.random.randint(0, 255, size=(2, 2, 1))
print(image)
```

```python
#灰度图
import numpy as np
import matplotlib.pyplot as plt

# 生成随机的灰度图像，每张图像可以用一个 28x28x1 的三维数组表示，其中第三个维度表示图像的通道数，图像是0-255之间，如果三个通道就是彩色图了
image = np.random.randint(0, 255, size=(28, 28, 1))
# 如果包含 60,000 张训练图像
# 整个数据集可以用一个形状为 (60000, 28, 28, 1) 的四维数组来表示，也就是60000行数据，28*行28列，1个通道。
# 展示灰度图像
plt.imshow(image.squeeze(), cmap="gray")
plt.show()
```
