```python
# 数据集的开源地址：https://github.com/google-research/mnist-c/
import os
import numpy as np
import matplotlib.pyplot as plt
#加载数据集并打印每个子文件夹前10个数据集
data_root = 'D:/test/tensorflow/dataset/mnist_c/mnist_c'
dirlist=os.listdir(data_root)
fig, axs = plt.subplots(len(dirlist), 10, figsize=(10, 10))

for i, folder_name in enumerate(dirlist):
    folder_path = os.path.join(data_root, folder_name)
    if os.path.isdir(folder_path):
        file_path = os.path.join(folder_path, 'train_images.npy')
        data = np.load(file_path)
        for j in range(0,10):
            axs[i, j].imshow(data[j].reshape(28,28), cmap='gray')
            axs[i, j].axis('off')
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
"""
  传入mnist-c，数据会非常大加载数据很慢
"""
def loadDataMnistC(data_root):
    g_train_images=np.empty((0, 28, 28, 1))
    g_train_labels=[]
    g_test_images=np.empty((0, 28, 28, 1))
    g_test_labels=[]
    dirlist=os.listdir(data_root)
    for i, folder_name in enumerate(dirlist):
        folder_path = os.path.join(data_root, folder_name)
        if os.path.isdir(folder_path):
            train_images = np.load(os.path.join(folder_path, 'train_images.npy'))
            g_train_images=np.vstack((g_train_images,train_images))
            train_labels = np.load(os.path.join(folder_path, 'train_labels.npy'))
            g_train_labels = np.concatenate((g_train_labels, train_labels))
            test_images = np.load(os.path.join(folder_path, 'test_images.npy'))
            g_test_images=np.vstack((g_test_images,test_images))
            test_labels = np.load(os.path.join(folder_path, 'test_labels.npy'))
            g_test_labels = np.concatenate((g_test_labels, test_labels))
    return (g_train_images,g_train_labels),(g_test_images,g_test_labels)
# 加载 MNIST-C 数据集
data_root = 'D:/test/tensorflow/dataset/mnist_c/mnist_c'
(x_train, y_train), (x_test, y_test) =loadDataMnistC(data_root)
x_train,y_train,x_test,y_test=np.array(x_train),np.array(y_train),np.array(x_test),np.array(y_test)
print(x_train.shape)
print(np.array(y_train).shape)
```
